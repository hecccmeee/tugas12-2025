## perbandingan nilai tatap muka vs daring

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'ID Siswa': ['Siswa-1', 'Siswa-2', 'Siswa-3', 'Siswa-4', 'Siswa-5'],
    'Jenis Pembelajaran': ['Daring', 'Tatap Muka', 'Daring', 'Tatap Muka', 'Daring'],
    'Nilai Ujian': [69, 75, 85, 63, 99],
    'Tingkat Kesulitan': [2, 5, 1, 1, 3],
    'Kehadiran (%)': [78.78, 61.36, 66.10, 76.79, 76.00]
}

df = pd.DataFrame(data)

daring = df[df['Jenis Pembelajaran'] == 'Daring']
tatap_muka = df[df['Jenis Pembelajaran'] == 'Tatap Muka']

plt.figure(figsize=(10, 6))

plt.plot(daring['ID Siswa'], daring['Nilai Ujian'], label='Daring', marker='o')
plt.plot(tatap_muka['ID Siswa'], tatap_muka['Nilai Ujian'], label='Tatap Muka', marker='o')

plt.xlabel('ID Siswa')
plt.ylabel('Nilai Ujian')
plt.title('Perbandingan Nilai Ujian: Tatap Muka vs Daring')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

plt.show()
```

## Distribusi Nilai Ujian Daring vs Tatap Muka

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'ID Siswa': ['Siswa-1', 'Siswa-2', 'Siswa-3', 'Siswa-4', 'Siswa-5',
                 'Siswa-6', 'Siswa-7', 'Siswa-8', 'Siswa-9', 'Siswa-10',

                ],
    'Jenis Pembelajaran': ['Daring', 'Tatap Muka', 'Daring', 'Tatap Muka', 'Daring',
                           'Daring', 'Daring', 'Tatap Muka', 'Daring', 'Tatap Muka',

                          ],
    'Nilai Ujian': [69, 75, 85, 63, 99,
                   74, 99, 83, 88, 62,

                  ]
}

df = pd.DataFrame(data)
df_daring = df[df['Jenis Pembelajaran'] == 'Daring']
df_tatap_muka = df[df['Jenis Pembelajaran'] == 'Tatap Muka']

plt.figure(figsize=(10, 6))
sns.histplot(df_daring['Nilai Ujian'], bins=10, kde=True, label='Daring', color='blue')
sns.histplot(df_tatap_muka['Nilai Ujian'], bins=10, kde=True, label='Tatap Muka', color='orange')
plt.xlabel('Nilai Ujian')
plt.ylabel('Frekuensi')
plt.title('Distribusi Nilai Ujian Daring vs Tatap Muka')
plt.legend()
plt.show()
```

## Distribusi Nilai Ujian

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Nilai Ujian'], bins=10, kde=True)
plt.title('Distribusi Nilai Ujian')
plt.xlabel('Nilai Ujian')
plt.ylabel('Frekuensi')
plt.show()

```

## Hubungan Antara Jenis Pembelajaran dan Nilai Ujian

```python
sns.boxplot(x='Jenis Pembelajaran', y='Nilai Ujian', data=df)
plt.title('Hubungan Antara Jenis Pembelajaran dan Nilai Ujian')
plt.show()
```

## Perbandingan Rata-rata Nilai Ujian

```python
average_scores = df.groupby('Jenis Pembelajaran')['Nilai Ujian'].mean().reset_index()

sns.barplot(x='Jenis Pembelajaran', y='Nilai Ujian', data=average_scores)
plt.title('Perbandingan Rata-rata Nilai Ujian')
plt.show()
```
